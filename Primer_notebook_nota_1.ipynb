# 1. Problema de ingeniería (ejemplo sencillo)

Un ingeniero agrícola necesita calcular **el área de un terreno rectangular** para estimar la cantidad de fertilizante que debe aplicar.

La fórmula es:

[
Área = largo \times ancho
]

---

# 2. Definición del algoritmo

Primero debemos identificar **Entradas, Proceso y Salidas**.

### Entradas (datos que recibe el programa)

* `largo` del terreno (metros)
* `ancho` del terreno (metros)

### Proceso (cálculo que realiza el algoritmo)

* Multiplicar largo por ancho

[
area = largo \times ancho
]

### Salida (resultado que entrega el programa)

* Área del terreno en **metros cuadrados**

---

# 3. Pseudocódigo

El pseudocódigo permite expresar la lógica **antes de escribir Python**.

```
Inicio

Leer largo
Leer ancho

area = largo * ancho

Mostrar area

Fin
```

---

# 4. Implementación en Python

```python
# Programa para calcular el área de un terreno rectangular
# Aplicación básica en ingeniería agrícola

# Entrada de datos
largo = float(input("Ingrese el largo del terreno en metros: "))
ancho = float(input("Ingrese el ancho del terreno en metros: "))

# Proceso
area = largo * ancho

# Salida
print("El área del terreno es:", area, "metros cuadrados")
```

---

# 5. Ejemplo de ejecución

# 6. Conceptos de Python utilizados

* **Variables:** `largo`, `ancho`, `area`
* **Operador matemático:** `*` (multiplicación)
* **Entrada de datos:** `input()`
* **Conversión numérica:** `float()`
* **Salida:** `print()`
